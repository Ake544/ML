```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('area-price.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.area,df.price, color='red', marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict(np.array(5000).reshape(-1, 1))
```

```python
d = pd.read_csv('area_to_predict.csv')
d.head(3)
```

```python
reg.predict(np.array(d).reshape(-1,1))
```

```python
p = reg.predict(np.array(d).reshape(-1,1))
d['Price'] = p
d.to_csv('area_to_predict.csv')
```

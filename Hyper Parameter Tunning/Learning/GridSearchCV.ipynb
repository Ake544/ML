```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df['flowers'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('flowers',axis='columns'),df.flowers) 
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=10, gamma='auto'), iris.data, iris.target)
```

```python
import numpy as np
```

```python
kernels = ['rbf','linear']
C = [1,10,20]
avg_score = {}
for kval in kernels:
   for cval in C:
        cv_score = cross_val_score(SVC(kernel=kval,C=cval, gamma='auto'), iris.data, iris.target)
        avg_score[kval + '_' + str(cval)] = np.average(cv_score)

avg_score
```

GridSerachCV


```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
}, cv=5, return_train_score=False)
clf.fit(iris.data, iris.target)
```

```python
df = pd.DataFrame(clf.cv_results_)
df.head()
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
clf.best_params_
```

```python
clf.best_score_
```

Randomized Serach CV


```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['rbf','linear']
},
cv=5,
return_train_score=False,
n_iter=2
)

rs.fit(iris.data, iris.target)
pd.DataFrame(rs.cv_results_)[['param_C','param_kernel','mean_test_score']]
```
